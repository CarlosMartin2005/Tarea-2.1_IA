```python
import pandas as pd

df = pd.read_csv('./altura_peso.csv')
```

```python
# Conjunto de datos de entrenamiento
datos = df[['altura', 'peso']]
```

```python
datos.head()
```

```python
import seaborn as sb

# sb.pairplot(datos)
sb.scatterplot(x='altura', y='peso', data=datos, hue='peso', palette='coolwarm')
```

```python
#Preparación de los datos
X = datos["altura"]
y = datos["peso"]
```

```python
#Procesamiento de los datos
X_train = X.values.reshape(-1, 1)
y_train = y.values.reshape(-1, 1)
```

```python
#Importar el modelo de regresión lineal
from sklearn.linear_model import LinearRegression

modelo = LinearRegression()
```

```python
#Entrenar el modelo
modelo.fit(X_train, y_train)
```

```python
#Intentar hacer predicciones, ingresando una altura, y que regrese el peso estimado
altura = 1.88
peso_estimado = modelo.predict([[altura]])
print(f"Para {altura} metros de altura, se estima un peso de {peso_estimado[0][0]} kg")
```

```python
modelo.score(X_train, y_train)
```

```python
# ¿Funciona bien o no? ¿Por qué cree que es así?
# El modelo no funciona bien en la mayoría de los casos, estuve probando con varios datos de altura reales 
# y el modelo no se ajusta a el peso real.
# Creo que es así porque la relación entre la altura y el peso no es lineal, y el modelo que se utilizó
# en este caso es de regresión lineal. Aún con esta desventaja, el procentaje de exactitud es del 82%.
```
